```python
class Calculator:
    def __init__(self):
        self.expression = []
        self.history = []   
        
    def display(self):
        expression_str = " ".join(self.expression) if self.expression else "0"
        print("\n" + "=" * 30)
        print(expression_str)
        print("=" * 30 + "\n")

    def reset(self):
        self.expression.clear()
        self.history.clear()
        print("Calculator reset. Current expression: 0")

    def back(self):
        if self.history:
            self.history.pop() 
            if self.history:
                self.expression = self.history[-1] 
            else:
                self.expression.clear()
            self.display()
        else:
            print("No operation to undo.")

    def add(self, value):
        self.history.append(self.expression.copy())
        self.expression.append(f"+ {value}")
        self.display()

    def subtract(self, value):
        self.history.append(self.expression.copy())
        self.expression.append(f"- {value}")
        self.display()

    def multiply(self, value):
        self.history.append(self.expression.copy())
        self.expression.append(f"* {value}")
        self.display()

    def divide(self, value):
        if value == 0:
            print("Math ERROR")
            use_back = input("Go back? (y/n): ").strip().lower()
            if use_back == 'y':
                self.back()
            else:
                print("Operation cancelled.")
        else:
            self.history.append(self.expression.copy())
            self.expression.append(f"/ {value}")
            self.display()

    def equals(self):
        if not self.expression:
            print("Current expression: 0")
            return

        try:
            expression_str = "".join(self.expression)
            result = eval(expression_str)
            print(f"Result: {result}")
            self.reset()
        except Exception as e:
            print(f"Error evaluating expression: {e}")

    def process_input(self, command):
        if command == "C" or command == "c":
            self.reset()
        elif command == "B" or command == "b":
            self.back()
        elif command.startswith("add"):
            value = float(command[4:])
            self.add(value)
        elif command.startswith("subtract"):
            value = float(command[9:])
            self.subtract(value)
        elif command.startswith("multiply"):
            value = float(command[9:])
            self.multiply(value)
        elif command.startswith("divide"):
            value = float(command[7:])
            self.divide(value)
        elif command == "=":
            self.equals()
        else:
            print("Invalid command.")

def main():
    print("Commands: add 1, subtract 2, multiply 3, divide 4, C to clear, B to back, =, or 'exit' to quit:")
    calc = Calculator()
    calc.display()

    while True:
        command = input("Enter command: ")
        if command.lower() == "exit":
            print("Exiting the calculator.")
            break
        calc.process_input(command)

if __name__ == "__main__":
    main()
```
