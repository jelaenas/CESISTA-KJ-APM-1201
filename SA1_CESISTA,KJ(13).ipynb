## Perform experimental analysis to test the hypothesis that Pythonâ€™s sorted method runs in O(nlogn) time on average.

```python
import time
import numpy as np
import matplotlib.pyplot as plt

def time_sort(n):
    array = np.random.randint(0, 100000000, size=n)
    
    start = time.time()
    sorted_array = sorted(array)
    end = time.time()
    
    total_time = end - start
    return total_time

def experiment(max_n, step):
    sizes = []
    times = []
    
    for n in range(step, max_n + 1, step):
        total_time = time_sort(n)
        sizes.append(n)
        times.append(total_time)
        print(f"Array of size {n} was sorted in {total_time:.6f} seconds.")

    return sizes, times

max_n = 10000000
step = 1000000

sizes, times = experiment(max_n, step)

hypothesis = [(n * np.log(n))/1e9 for n in sizes]

print("\nTiming Analysis:")
print(f"Min Time: {min(times):.6f} seconds")
print(f"Max Time: {max(times):.6f} seconds")

plt.figure(figsize=(10, 6))
plt.plot(sizes, times, color='blue', label='Experimental Sorting Time')
plt.plot(sizes, hypothesis, color='red', label='Expected O(n log n)')
plt.title('Sorted Method Time')
plt.xlabel('Array Size (n)')
plt.ylabel('Time (secs)')
plt.yscale('log')
plt.legend()
plt.show()
```
